## Prediting Heart Disease: Binary Classification Neural Network

#### Introduction: Packages and Data

```python
#import packages
import tensorflow as tf
from tensorflow import keras
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#initialize random seed to avoid repeating it throughout the code

keras.utils.set_random_seed(42)
```

```python
#import data
df = pd.read_csv("http://storage.googleapis.com/download.tensorflow.org/data/heart.csv")
##This data was made available online by Francois Chollet. 
```

```python
df.shape
```

```python
df.head()
```

```python
df.tail()
```

##### Data Preprocessing

```python
#group variables by type
categoricals = ["sex", "cp", "fbs", "restecg", "exang", "ca", "thal"]
numerics = ["age", "trestbps", "chol", "thalach", "oldpeak", "slope"]
```

```python
df= pd.get_dummies(df, columns= categoricals)
df.head()
```

```python
bool_cols = df.select_dtypes(include='bool').columns
df[bool_cols] = df[bool_cols].astype(int)
```

```python
df.head()
```

```python
#extract dependent variable
y = df.pop('target')
```

```python
#check out frequencies in the dependent variable
y.value_counts(normalize = True)
```

##### Split the Data into Train and Test

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df, y, test_size = 0.2, stratify =y)
```

```python
#let's find means and stds to use to normalize the data
means= x_train[numerics].mean()
sd = x_train[numerics].std()
means
```

```python
#normalize data using means and std
x_train[numerics]= (x_train[numerics]- means)/sd
x_test[numerics] =(x_test[numerics]- means)/sd
x_train.head()
```

```python
#before building a model in Tensorflow it is better to convert the panda dataframes to numpy arrays
x_train = x_train.to_numpy()
x_test = x_test.to_numpy()
```

##### Build the Model

```python
#create a variable to use in the input layer
num_columns = x_train.shape[1]

#input layer
input = keras.Input(shape=(num_columns,))

#hidden layer
h = keras.layers.Dense(16, activation = "relu", name = "Hidden")(input)

#output layer
output = keras.layers.Dense(1, activation = "sigmoid", name = "Output")(h)

#bind input output in the model
model = keras.Model(input, output)
```

```python
#see an overview of the model
model.summary()
```

```python
#let's visualize the network
keras.utils.plot_model(model, show_shapes = True)
```

##### Model Optimization

```python
#set parameters
model.compile(optimizer = "adam", loss = "binary_crossentropy", metrics = ["accuracy"])
```

##### Train the Model

```python

history = model.fit(x_train,
                    y_train,
                    epochs=300,
                    batch_size=32,
                    validation_split=0.2)
```

##### Check Model for Overfitting

```python
history_dict = history.history
history_dict.keys()
```

```python
plt.clf()
loss_values = history_dict["loss"]
val_loss_values = history_dict["val_loss"]
epochs = range(1, len(loss_values)+1)
plt.plot(epochs, loss_values, "bo", label = "Training loss")
plt.plot(epochs, val_loss_values, "b", label = "Validation loss")
plt.title("Training and Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

```python
#let's check out the accuracy by plotting it
plt.clf()
acc = history_dict["accuracy"]
val_acc = history_dict["val_accuracy"]
plt.plot(epochs, acc, "bo", label = "Training Accuracy")
plt.plot(epochs, val_acc, "b", label = "Validation Accuracy")
plt.title("Training and Validation Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

##### Model Evaluation

```python
model.evaluate(x_test, y_test)
```

It does better than a baseline model, but not by much. I will go on to try different optimizer paramters to see if it can improve.
